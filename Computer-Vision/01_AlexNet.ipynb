# 2. AlexNet: A Breakthrough in Deep Learning

[AlexNet](https://proceedings.neurips.cc/paper_files/paper/2012/file/c399862d3b9d6b76c8436e924a68c45b-Paper.pdf), a Convolutional Neural Network architecture, made history by winning the ImageNet Challenge of 2012 with a top-5 error rate of 16.4%. This was a substantial improvement over the previous best of 25.8%, marking a significant milestone in the field of deep learning.

The mastermind behind AlexNet was **Alex Krizhevsky**, a computer scientist and researcher. Working with Ilya Sutskever and Geoffrey Hinton at the University of Toronto, Krizhevsky developed AlexNet, which introduced several key innovations in deep learning, including:

- **Deep Convolutional Networks:** AlexNet employed a deeper architecture than previous models, demonstrating the effectiveness of stacking multiple layers.

- **ReLU Activation Functions:** The use of ReLU activation functions helped to address the vanishing gradient problem, enabling faster training.

- **Dropout Regularization:** Dropout was introduced to prevent overfitting by randomly dropping neurons during training.

AlexNet's success had a profound impact on the field of deep learning, inspiring further research and development in computer vision and other domains. As of 2024, AlexNet has been cited over 164,083 times, solidifying its position as one of the most influential papers in the history of deep learning.

# 3. AlexNet Architecture

AlexNet architecture consists of a total of eight layers: five convolutional layers (three of which are followed by max pooling layers) and three fully connected layers. The final layer is made up of 1000 units (corresponding to the 1000 classes in the ImageNet dataset) and uses a softmax activation function.

AlexNet was similar to [LeNet-5](https://yann.lecun.com/exdb/publis/pdf/lecun-01a.pdf), a model introduced earlier in the 1990s, but with a deeper architecture and the use of ReLU activation functions in the hidden layers instead of sigmoid functions.

To address overfitting, AlexNet was the first paper to introduce the dropout regularization technique. Dropout remains one of the most effective regularization techniques used in deep learning today.

## AlexNet Architecture Breakdown

The AlexNet architecture takes an input image of size (3, 224, 224) or (224, 224, 3).

1. **Convolutional Layer 1:**
    - Uses 96 filters (split across two GPUs: 48 + 48).
    - Each filter is 11x11 in size.
    - Has a stride of 4.
    - Produces feature maps of size 54x54.

2. **Max Pooling Layer 1:**
    - Has a pool size of 3.
    - Has a stride of 2.

3. **Convolutional Layer 2:**
    - Uses 256 filters.
    - Each filter is 5x5 in size.

4. **Convolutional Layers 3, 4, and 5:**
    - Use 384, 384, and 256 filters, respectively.
    - Each filter is 3x3 in size.
    - No max pooling layers.

5. **Max Pooling Layer 2:**
    - Has a pool size of 3.
    - Has a stride of 2.

6. **Flatten Layer:**
    - Converts the feature maps into a 1D vector.
    - **Note:** This layer has no learnable parameters, as it simply reshapes the data.

7. **Fully Connected Layers:**
    - Two layers with 4096 units each.
    - Final layer with 1000 units (corresponding to the 1000 classes in ImageNet) and a softmax activation function.

**Activation Function:**

- AlexNet uses ReLU activation functions in all hidden layers, which is faster than sigmoid functions and helps to prevent the vanishing gradient problem.

# 4. Reducing Overfitting: Regularization Techniques in AlexNet

To mitigate overfitting, AlexNet employed two key regularization techniques: data augmentation and dropout.

**Data Augmentation**

- **Purpose:** To expand the training dataset by creating artificial variations of existing images.

- **Methods:**
    - Horizontal flipping: Randomly mirroring images horizontally.
    - Color intensity changes: Adjusting the brightness, contrast, or saturation.

**Dropout**

- **Purpose:** To make the network more robust by randomly dropping neurons during training.

- **Mechanism:**
    - Neurons are randomly dropped with a probability of 0.5 in AlexNet's first two fully connected layers.
    - Dropped neurons are not considered during forward and backward propagation.
    - This forces the network to learn without relying on specific neurons, improving its generalization ability.
- **Benefits:**
    - Can be seen as a computationally efficient form of ensemble learning, as it creates multiple networks with shared weights.

By using these regularization techniques, AlexNet was able to achieve better performance and generalization on the ImageNet dataset.
